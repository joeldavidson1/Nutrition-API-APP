# Proximates

```python
import pandas as pd
df_proximates = pd.read_excel('COFID_dataset.xlsx', sheet_name='1.3 Proximates')
df_proximates
```

```python
df_proximates = df_proximates.drop([0, 1])
df_proximates = df_proximates.drop(df_proximates.columns[[4, 6, 8, 15, 26, 28, 29, 30, 31, 31, 32, 33, 34, 36, 37, 38, 40, 41, 42, 43, 44]] , axis=1)
df_proximates
```

```python
# Rename columns to remove duplications
df_proximates.rename({'Energy (kcal) (kcal)' : 'Energy (kcal)', 'Energy (kJ) (kJ)': 'Energy (kJ)'}, axis=1, inplace=True)
df_proximates
```

# Inorganics

```python
df_inorganics = pd.read_excel('COFID_dataset.xlsx', sheet_name='1.4 Inorganics')
df_inorganics
```

```python
# Add missing column name to first column
df_inorganics.rename(columns={' ': 'Food Code'}, inplace=True)
```

```python
df_inorganics = df_inorganics.drop([0, 1])
df_inorganics = df_inorganics.drop(columns=['Description', 'Previous', 'Main data references', 'Footnote'])
df_inorganics
```

# Vitamins

```python
df_vitamins = pd.read_excel('COFID_dataset.xlsx', sheet_name='1.5 Vitamins')
df_vitamins
```

```python
df_vitamins = df_vitamins.drop([0, 1])
df_vitamins = df_vitamins.drop(columns=['Description', 'Previous', 'Main data references', 'Footnote', 'Retinol Equivalent (µg)', 'Niacin equivalent (mg)'])
df_vitamins
```

# Merging the dataframes 

```python
df_all = df_proximates.merge(df_inorganics)
df_all = df_all.merge(df_vitamins)
df_all
```

```python
# df_all = df_all.fillna("N/A")
```

# Removing duplicates

```python
df_all['Food Code'].is_unique
```

```python
df_all[df_all.duplicated(['Food Code'], keep=False)]
```

### Change one of the food codes to one that doesn't exist

```python
new_food_code = '13-000'
df_all[df_all['Food Code'] == new_food_code]
```

```python
df_all.at[51, 'Food Code'] = new_food_code
```

```python
df_all['Food Code'].is_unique
```

# Removing food items with Group = 'BTM' as this was not provided in the CoFID PDF

```python
df_all[df_all['Group'] == 'BTM']
```

```python
df_all.drop(df_all[df_all['Group'] == 'BTM'].index, inplace=True)
```

# Trace (Tr) amounts and N

```python
import numpy as np

# Replace 'Tr' with a small positive value
df_all.replace('Tr', 0.01, inplace=True)

# Replace 'N' with NaN
df_all.replace('N', np.nan, inplace=True)

# Replacing blank spaces with NaN
df_all.replace(r'^\s+$', pd.notna, regex=True, inplace=True)

# df_all.fillna(np.nan)
```

# Dataframe to CSV file

```python
df_all.to_csv('combined_nutrition_tables.csv', index=False, na_rep='NULL')
```
